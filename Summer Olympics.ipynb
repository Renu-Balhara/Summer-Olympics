```python
import pandas as pd
df = pd.read_csv('summer.csv')
```

# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
print("Number of cities Summer Olmpics is held so far ->", len(df['City'].unique()))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df1 = df[df['Medal'] == 'Gold']
ax = df1.groupby('Sport').count()['Medal'].sort_values(ascending=False).head().plot(x='Sport', y='Medals', 
                                                    kind='bar', figsize=(6,5), title="Sports with most number of Gold medals")
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x(), p.get_height()))
ax.legend(['Gold Medals'])
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
ax = df.groupby('Sport').count()['Medal'].sort_values(ascending=False).head().plot(x='Sport', y='Medals', kind='bar', 
                                            figsize=(5,5), title="Sports with most number of Medals")
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x(), p.get_height()))
ax.legend(['Medals'])
```

### 4. Which player has won most number of medals? (Top 5)

```python
ax = df.groupby('Athlete').count()['Medal'].sort_values(ascending=False).head().plot(x='Athlete', y='Medals', 
                                                    kind='bar', figsize=(5,5), title="Athletes with most number of Medals")
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x(), p.get_height()))
ax.legend(['Medals'])
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
df1 = df[df['Medal'] == 'Gold']
ax = df1.groupby('Athlete').count()['Medal'].sort_values(ascending=False).head().plot(x='Athlete', y='Medals', 
                                                    kind='bar', figsize=(5,5), title="Athletes with most number of Gold medals")
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x(), p.get_height()))
ax.legend(['Gold Medals'])
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
df1 = df[df['Country'] == 'IND']
df1 = df1[df1['Medal'] == 'Gold']
df2 = df1.sort_values(by='Year', ascending=True).head(1)
for year in df2['Year']:
    print("India won it's first Gold Medal in Summer Olympics in", year)
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
ax = df.groupby('Event').count()['Athlete'].sort_values(ascending=False).head().plot(x='Event', y='Players', 
                                            kind='bar', figsize=(6,6), title="Event with most number of players")
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x(), p.get_height()))
ax.legend(['Players'])
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
df1 = df[df['Gender'] == 'Women']
df1 = df1[df1['Medal'] == 'Gold']
ax = df1.groupby('Sport').count()['Medal'].sort_values(ascending=False).head().plot(x='Sport', y='Medals', 
                                            kind='bar', figsize=(5,5), title="Sport with most number of female Gold medalists")
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x(), p.get_height()))
ax.legend(['Gold Medals'])
```
